```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, widgets
from IPython.display import display, HTML

# Example dataset
data = {
    "Movie": ["Inception", "The Dark Knight", "Interstellar", "Avatar", "Titanic", "The Godfather", "The Shawshank Redemption"],
    "Rating": [8.8, 9.0, 8.6, 7.8, 7.9, 9.2, 9.3]
}
df = pd.DataFrame(data)

# Function to convert rating into star symbols
def rating_to_stars(rating):
    full_stars = int(rating)  # whole stars
    return "⭐" * full_stars

# Function to display movies above a chosen rating
def show_movies(min_rating):
    filtered = df[df["Rating"] >= min_rating].copy()
    filtered["Stars"] = filtered["Rating"].apply(rating_to_stars)  # add stars column

    # Display styled table
    display(HTML(filtered[["Movie", "Rating", "Stars"]].sort_values(by="Rating", ascending=False).to_html(index=False)))

    # Plot bar chart
    plt.figure(figsize=(8, 4))
    plt.bar(filtered["Movie"], filtered["Rating"], color="orange", edgecolor="black")
    plt.xticks(rotation=45)
    plt.ylabel("Rating")
    plt.title(f"Movies with Rating ≥ {min_rating}")
    plt.show()

# Custom slider with style
slider = widgets.IntSlider(
    value=8,
    min=0,
    max=10,
    step=1,
    description="Min Rating:",
    style={'description_width': 'initial'},
    continuous_update=False
)

interact(show_movies, min_rating=slider);
```
